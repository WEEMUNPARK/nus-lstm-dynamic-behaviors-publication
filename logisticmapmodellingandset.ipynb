```python
import numpy as np
def logisticmapmodel(r):
    n=20000
    li=[]
    x=np.random.random()
    x_1=x*(1-x)*r
    li.append(x)
    li.append(x_1)
    while n>0:
        x=x_1
        x_1=x*(1-x)*r
        li.append(x_1)
        n-=1
    return li
```

```python
y=[]
x=[]
for r in range(150,400):
    y.append(logisticmapmodel(r/100))
    x.append([r/100])
for j in range(len(x)):
    x[j]=x[j]*len(logisticmapmodel(2))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plt.ylim(0,1)
plt.plot(x, y, '^', color='blue', alpha=0.4, markersize = 0.013)
plt.title('logistic map')
plt.xlabel('r')
plt.ylabel('x')
plt.show()
```
